```python
import numpy as np
import matplotlib.pyplot as plt

a = -1
b = 4
c = 0.4
# ax + by + c = 0
# y = (-ax-c)/b
def plotline(a, b, c):
  x = np.linspace(-2, 4, 50)
  y = (-a*x-c)/b
  plt.axvline(0, -1, 1, color='k', linewidth=1)
  plt.axhline(0, -2,41, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid(True)

plotline(a, b, c)
```

```python
p1 = (2, 0.4)
retorno = a*p1[0] + b*p1[1] + c

p2 = (1, 0.6)

p3 = (1, -0.4)

plotline(a, b, c)
plt.plot(p1[0], p1[1], marker='o', color='b')
plt.plot(p2[0], p2[1], marker='o', color='y')
plt.plot(p3[0], p3[1], marker='o', color='r')
```

```python
ret1 = a*p1[0] + b*p1[1] + c
ret2 = a*p2[0] + b*p2[1] + c
ret3 = a*p3[0] + b*p3[1] + c

print("retorno 1:","%.2f" % ret1)
print("retorno 2: ","%.2f" % ret2)
print("retorno 3: ","%.2f" % ret3)
```

```python
from sklearn.datasets import *

np.random.seed(46)

X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1,
                           n_clusters_per_class=1)


plt.scatter(X[:,0], X[:,1], marker='o', c=Y)

p= X[10]
plt.plot(p[0], p[1], marker='^', markersize=20)

def plotmodel(w1, w2, b):
  x = np.linspace(-2, 4, 50)
  y = (-w1*x-b)/w2
  plt.axvline(0, -1, 1, color='k', linewidth=1)
  plt.axhline(0, -2,41, color='k', linewidth=1)
  plt.plot(x, y)
  plt.grid(True)

w1 = 5 # a
w2 = 1 # b
b = -0.4 # c

plotmodel(w1, w2, b)

p = (-1, 1)
print(w1*p[0] +w2*p[1]+ b)
```

```python
def classify(p, w1, w2, b):
  ret = w1*p[0] + w2*p[1] + b
  if ret >=0:
    return 1, 'yellow'
  else:
    return 0, 'blue'

p = (-2, -1)
classe, cor = classify(p, w1, w2, b)
print(classe, cor)

plotmodel(w1, w2, b)
plt.plot(p[0], p[1], marker='^',color=cor, markersize=20 )
```
